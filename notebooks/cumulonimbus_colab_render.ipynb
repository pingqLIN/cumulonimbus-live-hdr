# Cumulonimbus Live HDR Colab Render

This notebook runs the repository's non-interactive Colab render entrypoint. Use it for batch output, metrics, and artifact generation. It is not intended to host the public website or provide live per-visitor rendering.

```python
# Optional: mount Drive when you want artifacts to survive runtime shutdown.
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Configure repository source.
# For a private repo, provide a token through Colab secrets or upload a zip to Drive instead.
REPO_URL = 'https://github.com/pingqLIN/cumulonimbus-live-hdr.git'
BRANCH = 'codex/colab-port'
WORKDIR = '/content/cumulonimbus-live-hdr'
ARTIFACT_DIR = '/content/drive/MyDrive/cumulonimbus-live-hdr/colab-outputs'
```

```python
import os, shutil, subprocess, textwrap

def run(command, cwd=None):
    print('\n$ ' + command)
    subprocess.run(command, shell=True, cwd=cwd, check=True)

if os.path.exists(WORKDIR):
    shutil.rmtree(WORKDIR)
run(f'git clone --depth 1 --branch {BRANCH} {REPO_URL} {WORKDIR}')
run('node --version')
run('npm --version')
```

```python
# Quick smoke render. Increase width/height/seconds after this succeeds.
run('npm run colab:render -- --mode quick --install --check -- --width 360 --height 640 --seconds 1', cwd=WORKDIR)
```
